### Creating a Series

```python
s = pd.Series([10, 20, 30])
s
```

The values in the index are actually labels to the data, to help you retrieve it later.

You can create more appropriate labels.

```python
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s
```
