Apply filter to a column

```python
df['VALVE_VSL_VC'] == 'VSL-4'
```

The result may be interpreted as a mask, which later you can apply to the dataframe

Create a filter (or a mask)

```python
vsl_filter = (df['VALVE_VSL_VC'] == 'VSL-4')
```

And then apply the filter

```python
df[vsl_filter]
```

It also works with the 'loc' method

```python
df.loc[vsl_filter]
```

Which allows you to apply additional filters

```python
df.loc[vsl_filter, 'PART_CODE_SHORT']
```

```python
set(df.loc[vsl_filter, 'PART_CODE_SHORT'])
```

You may apply more filters at once

```python
my_filter = (df['VALVE_VSL_VC'] == 'VSL-4') & (df['PART_CODE_SHORT']=='Axial')
```

```python
df.loc[my_filter]
```

& = AND

| = OR

~ = inverts the filter

```python
my_filter = ~(df['VALVE_VSL_VC'] == 'VSL-4') & (df['PART_CODE_SHORT']=='Axial')
```

```python
df.loc[my_filter]
```

Another feature is to search within a list of values

```python
set(df['VALVE_MIN_TEMP_VC'])
```

```python
low_temp = ('-100C', '-50C', '-46C')
filt = df['VALVE_MIN_TEMP_VC'].isin(low_temp)
```

```python
df.loc[filt, 'VALVE_MIN_TEMP_VC']
```

You may combine string methods


```python
my_filt = df['COMMODITY_CODE'].str.contains('VGA-25')
```

```python
df.loc[my_filt]
```
