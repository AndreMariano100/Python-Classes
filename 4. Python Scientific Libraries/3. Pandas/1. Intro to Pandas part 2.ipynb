# Access information within the database

## Access columns (series)

```python
df['SPECS']
```

```python
set(df['SPECS'])
```

```python
type(df['SPECS'])
```

## Instead of using the brackets (as in a dictionary) you may access like an attribute

```python
df.SPECS
```

```python
set(df.SPECS)
```

## You may access many columns at once

```python
df[['SPECS', 'PART_CODE_SHORT']]
```

```python
df.columns
```

## To access rows use the iloc, instead of using the dictionary notation

```python
df.iloc[0]
```

```python
df.iloc[1000]
```

## You may access many rows at once

```python
df.iloc[[10, 11]]
```

## You may access rows and columns at the same time 
### (iloc requires column number instead of column label)

```python
df.iloc[[10, 11], [0, 1, 3, 4]]
```

```python
df.iloc[0:10, [0, 1, 3, 4]]
```

## You may also use 'loc' instead of 'iloc'

```python
df.iloc[0]
```

```python
df.loc[0]
```

## With 'loc' you may call the columns by the appropriate labels

```python
df.loc[[0], ['SPECS', 'COMMODITY_CODE', 'PART_CODE']]
```

```python
df.loc[0:10, ['SPECS', 'COMMODITY_CODE', 'PART_CODE']]
```

```python
df.columns
```

```python
df.loc[0:10, 'SPECS': 'PART_CODE']
```
