## Using other information instead of the index

This is a temporary change


```python
df.set_index('COMMODITY_CODE')
```

The dataframe is still unchanged

```python
df
```

This is a permanent change to the DF

```python
df.set_index('COMMODITY_CODE', inplace=True)
```

```python
df
```

Now you can look by tags, instead os index numbers

```python
df.loc['VES-600-2-C23G']
```

With tags as index, 'loc' wont work

```python
df.loc[0]
```

But the iloc still works

```python
df.iloc[0]
```

You can revert the change to the index

```python
df.reset_index(inplace=True)
```

```python
df
```
