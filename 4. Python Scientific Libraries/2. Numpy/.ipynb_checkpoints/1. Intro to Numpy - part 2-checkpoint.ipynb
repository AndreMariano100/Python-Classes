## Creating arrays

```python
import numpy as np
```

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
a
```

```python
b = np.array(((1, 2, 3), (4, 5, 6)))
b
```

```python
c = np.array(('a', 'b', 'c'))
c
```
