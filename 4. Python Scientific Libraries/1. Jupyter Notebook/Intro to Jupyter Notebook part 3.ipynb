```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(style="darkgrid")
```

```python
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
angular_coeff = 1.5
```

```python
linear_coeff = 3
```

```python
x = np.linspace(-10, 10, 100)
```

```python
y = angular_coeff * x + linear_coeff
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```
